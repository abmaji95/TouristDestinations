```python
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
import re
import time
```

```python
driver = webdriver.Chrome(executable_path='C:/Users/A263741/Desktop/chromedriver_win32/chromedriver.exe')
seed_url = "https://www.tripadvisor.in"
driver.get(seed_url)
time.sleep(10)
soup = BeautifulSoup(driver.page_source, 'html.parser')

city_key_lst=['-g187147-Paris_Ile_de_France','-g304554-Mumbai_Maharashtra','-g186338-London_England']

Hotel = []
Hotel_url=[]
for name in soup.find_all('a',{"href":"/Hotels"}):
    Hotel.append(name.text.strip())

for i in city_key_lst:
    Hotel_path=seed_url+'/'+Hotel[0]+i+'-Hotels.html'
    Hotel_url.append(Hotel_path)

hotel = []
ratings = []
reviews = []

for j in Hotel_url:
    driver.get(j)
    time.sleep(10)
    soup1 = BeautifulSoup(driver.page_source, 'html.parser')
        
    
    for name in soup1.find_all('div',attrs = {"class":"listing_title"}):
        hotel.append(name.text.strip())
    
    
    for rating in soup1.find_all('a',attrs = {'class':'ui_bubble_rating'}):
        ratings.append(rating['alt'])

    
    for review in soup1.find_all('a',{'class':'review_count'}):
        reviews.append(review.text.strip())
    
print(hotel)
print(ratings)
print(reviews)

driver.quit()
```

```python
import pandas as pd
d1 = {'Hotel':hotel,'Ratings':ratings,'No_of_Reviews':reviews}
print(d1)
df = pd.DataFrame.from_dict(d1)
```
